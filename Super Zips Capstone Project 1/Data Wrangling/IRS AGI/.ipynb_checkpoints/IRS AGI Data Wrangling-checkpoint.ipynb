```python
import pandas as pd
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# read 2017 sample data filCSV file into dataframe
df = pd.read_csv('2017 IRS AGI Raw Data Subset Sample.csv', 
                     dtype={'STATE': 'category', 'ZIPCODE':np.int32, 'AGI_STUB':np.int32, 'N1':np.int32})

# zipsum_df: dataframe calculating Total No. of Returns per zip code
total_return_zip_df = pd.DataFrame(df.groupby('ZIPCODE').sum(),columns=['N1']) 

# merge total_return_zip_df to main df by zip code
zip_df = df.merge(total_return_zip_df, on='ZIPCODE') 

# create and calculate weighted avg Score column
zip_df['Score'] = (zip_df.N1_x / zip_df.N1_y) * zip_df.AGI_STUB 

# sum all weighted terms per zipcode
score_per_zip_df = zip_df.groupby('ZIPCODE').sum() 

# zip_state_only: DataFrame with State and Zipcode pairs only
zip_state_only = df[df.ZIPCODE != 0][['ZIPCODE','STATE']].drop_duplicates(subset='ZIPCODE') 

# score_per_state_zip_df: finalized DataFrame for analysis; created by merging zip_state_only and score_per_zip_df
score_per_state_zip_df = score_per_zip_df.merge(zip_state_only, on='ZIPCODE')[['ZIPCODE','STATE','Score']] # merge 

# display info check
score_per_state_zip_df.info()
```

```python
score_per_state_zip_df.Score.hist(bins=200)
plt.xlabel('Score')
plt.ylabel('Frequency')
```

```python
# Create histograms for Income score data by Zip Code aggregated by State (takes about 5 minutes to load)
histGrid = sns.FacetGrid(score_per_state_zip_df, col = 'STATE', col_wrap = 5, height = 2.25, hue='STATE', palette = 'bright')
histGrid = histGrid.map(sns.swarmplot,'Score',order=list(score_per_state_zip_df.STATE.unique())).set_axis_labels("Score")
```
