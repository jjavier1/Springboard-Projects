```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


mainDF = pd.read_csv('Census Main Analysis Data.csv', 
                     dtype={"ZIPCODE":np.int32,"STATE":'category',
                             'ZipScore':np.float32,'YEAR':np.int32,
                             'StateScore':np.float32}).iloc[:,1:]

# mainDF[['STATE','YEAR','StateScore']].drop_duplicates().pivot(index='STATE',columns='YEAR',values='StateScore')
mainDF.info()
```

```python
# Analyzing Education Scores of all 51 states
g = sns.FacetGrid(mainDF, col="STATE", col_wrap=5, height=2.25, hue='STATE', palette = 'bright')
g = g.map(plt.plot, "YEAR", "StateScore", marker=".")
```

```python
# Analyzing zip codes of the "state" with the best Income score trend, District of Columbia (DC)
g = sns.FacetGrid(mainDF[mainDF.STATE == 'DC'], col="ZIPCODE", col_wrap=5, height=2.25, hue='ZIPCODE', palette = 'bright')
g = g.map(plt.plot, "YEAR", "ZipScore", marker=".")
```

```python
# Analyzing zip codes of the state with the worst Income score trend, Mississippi (MS)
g = sns.FacetGrid(mainDF[mainDF.STATE == 'MS'], col="ZIPCODE", col_wrap=5, height=2.25, hue='ZIPCODE', palette = 'bright')
g = g.map(plt.plot, "YEAR", "ZipScore", marker=".")
```
