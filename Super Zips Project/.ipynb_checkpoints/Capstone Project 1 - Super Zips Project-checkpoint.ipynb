```python
import pandas as pd
import numpy as np
path = 'ACSST5Y2018.S1501_data_with_overlays_2020-05-02T194012.csv'
cols_list = ['GEO_ID','NAME','S1501_C01_016E', 'S1501_C01_016M', 'S1501_C01_017E', 
             'S1501_C01_017M', 'S1501_C01_018E', 'S1501_C01_018M']
zips_df = pd.read_csv(path, usecols=cols_list).drop(0)
# zips_df[zips_df['S1501_C01_016M']=='*****']
zips_df = zips_df.replace({'*****':0})
zips_df = zips_df.astype({'S1501_C01_016E': 'int32','S1501_C01_016M': 'int32',
                'S1501_C01_017E': 'int32','S1501_C01_017M': 'int32',
                'S1501_C01_018E': 'int32','S1501_C01_018M': 'int32',
               })
zips_df.info()
```
